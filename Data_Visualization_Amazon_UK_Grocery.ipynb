We see that the dataset has a lot of empty cells and cells with NaN values. To get a better picture we will create a function to give us the percentage of empty cells and cells with NaN values for each column.

```python
def percent_null(column):
  '''
  Iterate through all columns to get the percentage
  of each column or just pass one column to get percentage
  for that column.
  '''
  total_cells = np.prod(column.shape)
  missing_values = column.isnull().sum()

  percent_null = (missing_values / total_cells) *100

  return percent_null
```

```python
# Iterating through all the columns to get the percentage for each
# column by applying the function above (percent_null).
for column in amazon_dataset_raw.columns:
  percent_null_value = percent_null(amazon_dataset_raw[column])
  print(f"{column}:{percent_null_value:.2f}%")
```

## Data Analysis and Visualization

### Top performing brands.

We will start our data analysis and visualization with finding out top performing brands. Please note that this is done before data cleaning. Therefore, we might not get an accurate picture of our top brands as there may be spelling mistakes or error while entering brands in our dataset.

```python
# Top brands according to transactions.
top_brands = amazon_dataset_raw['Brand'].value_counts() # Counting the  number of transactions for each brand.
top_brands = top_brands[:10] # Taking the only top 10 brands.
top_brands = pd.DataFrame(top_brands)
top_brands
```

```python
# Top brands according to revenue
# Grouping the dataset by Brand column and then looking at the sum or their revenue in descending order.
top_brands_revenue = amazon_dataset_raw.groupby('Brand')['Revenue'].sum().sort_values(ascending=False)
top_brands_revenue = top_brands_revenue[:10] # Taking the only top 10 brands.
top_brands_revenue = pd.DataFrame(top_brands_revenue) # Converting it into a dataframe.
top_brands_revenue
```

We are selecting top 5 brands from transactions and Revenue (Pukka Herbs being common) to see the average transaction price comparison and understand little about their strategy.

```python
group_brands_avg_price = amazon_dataset_raw.groupby('Brand')
brands = []
avg_price = []

# Listing the groups for which we need the average transaction price.
groups = ['MORRISONS', 'Alpro', 'Tea Pigs', 'Twinings', 'Pukka Herbs', 'Tassimo', 'Lavazza', 'STARBUCKS', 'Lindt']

# Iterating through all groups defined above and getting the mean of price column for each group.
for group in groups:
    brands.append(group)
    avg_price.append(group_brands_avg_price.get_group(group)['Price'].mean())

brands_avg_price  = pd.DataFrame(avg_price, brands, columns=['Avg Transaction Price']) # Converting brands_avg_price into a dataframe.

brands_avg_price = brands_avg_price.sort_values('Avg Transaction Price', ascending=False) # Arranging the dataframe in descending order by Avg Transaction Price column.
brands_avg_price
```

```python
# Top brands according to unit sold.
# Grouping the dataset by Brand column and then looking at the sum of their unit sold (Sales) in descending order.
top_brands_unit_sold = amazon_dataset_raw.groupby('Brand')['Sales'].sum().sort_values(ascending=False)
top_brands_unit_sold = top_brands_unit_sold[:10] # Taking the only top 10 brands.
top_brands_unit_sold = pd.DataFrame(top_brands_unit_sold) # Converting the variable into a dataframe.
top_brands_unit_sold
```

Now that we have the data for the top brands according to number of transactions, Revenue and Unit Sold and also have the comparison of avg transaction price among the top brands, we will proceed to plot them on bar graphs to understand the dynamics.

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Top Brands') # Setting figure title.

sns.barplot(data=top_brands, x=list(top_brands.index), y=list(top_brands.iloc[:, -1]), hue=list(top_brands.index), ax=ax[0, 0]) # Plotting the data in bar graph.
ax[0, 0].set_title('Top Brands by # of Transactions') # Setting plot title.
ax[0, 0].tick_params(axis='x', rotation=45) # Changing tick parameters (x axis) for better visualization.
ax[0, 0].set_xlabel('Brands') # Setting label for x axis.
ax[0, 0].set_ylabel('Transactions')  # Setting label for y axis.

# Using EngFormatter with zero decimal places to display values in thousands (K) and millions (M) for improved readability
# and comprehension.
fmt = EngFormatter(places=0) # Creating an instance in the EngFormatter class.
# Plotting the data in bar graph.
sns.barplot(data=top_brands_revenue, x=list(top_brands_revenue.index), y=(top_brands_revenue.iloc[:, -1]), hue=list(top_brands_revenue.index), ax=ax[0, 1])
ax[0, 1].set_title('Top Brands by Revenue') # Setting plot title.
ax[0, 1].ticklabel_format(axis='y', style='plain')
ax[0, 1].yaxis.set_major_formatter(fmt) # Using EngFormatter on y axis.
ax[0, 1].tick_params(axis='x', rotation=45) # Changing tick parameters (x axis) for better visualization.
ax[0, 1].set_xlabel('Brands') # Setting label for x axis.
ax[0, 1].set_ylabel('Revenue') # Setting label for y axis.

# Plotting the data in bar graph.
sns.barplot(data=brands_avg_price, x=list(brands_avg_price.index), y=list(brands_avg_price.iloc[:, -1]), hue=list(brands_avg_price.index), ax=ax[1, 0])
ax[1, 0].set_title('Top Brands by Avg Transaction Price') # Setting plot title.
ax[1, 0].tick_params(axis='x', rotation=45) # Changing tick parameters (x axis) for better visualization.
ax[1, 0].set_xlabel('Brands') # Setting label for x axis.
ax[1, 0].set_ylabel('Avg Transaction Price') # Setting label for y axis.

# Plotting the data in bar graph.
sns.barplot(data=top_brands_unit_sold, x=list(top_brands_unit_sold.index), y=list(top_brands_unit_sold.iloc[:, -1]), hue=list(top_brands_unit_sold.index), ax=ax[1, 1])
ax[1, 1].set_title('Top Brands by Units Sold') # Setting plot title.
ax[1, 1].tick_params(axis='x', rotation=45) # Changing tick parameters (x axis) for better visualization.
ax[1, 1].set_xlabel('Brands') # Setting label for x axis.
ax[1, 1].set_ylabel('Units Sold') # Setting label for y axis.


plt.tight_layout()
plt.show()
```

Analysing these graphs provides valuable insights into the dynamics of the Amazon UK Grocery market. Conversely, Lavazza stands out with both a high transaction volume (Fig: Top Brands by # of Transactions) and a significant market share (Fig: Top Brands by Revenue). A deeper examination of the Avg Transaction Price and Unit Sold charts offers further clarity. ### String Matching

Stripping the strings (brand names) of spaces and converting them into lower case.

```python
amazon_dataset_clean['Brand'] = amazon_dataset_clean['Brand'].str.lower()
amazon_dataset_clean['Brand'] = amazon_dataset_clean['Brand'].str.strip()
```

Now that we have stripped the strings of the brand names of spaces and converted them into lowercase, we will create a function to match each string with every other string and store the matches in close matches, provided that the match percentage is over 90%. We will utilize Fuzzywuzzy for this task.

```python
def get_matches(df, column, min_ratio=90):
    '''
    Pass the dataset and the relevant column as argument
    to match a string with every other string in the column to
    find close matches (over 90%)
    '''
    strings = df[column].unique() # Getting unique strings from the specified column of the DataFrame

    # Iterating through unique strings.
    for i in strings:
        # Find matches for the current string in iteration against all other strings using token set ratio.
        matches = fuzzywuzzy.process.extract(i, strings, scorer=fuzzywuzzy.fuzz.token_set_ratio)
        # Filter matches based on minimum ratio and excluding the current string in iteration.
        close_matches = [match for match in matches if match[1] >= min_ratio and match[0] != i]
        # If close matches (over 90%) were found, we will print the current string in iteration
        # and iterate through close matches to print the close match and the ratio of match.
        if close_matches:
            print(f"Close matches for '{i}':")
            for match in close_matches:
                print(f"- {match[0]} (Match Ratio: {match[1]})")
```

```python
matches = get_matches(df=amazon_dataset_clean, column='Brand') # Finding matches in Brand column through the function created above.
matches
```
